```python
data={'department':['HR','IT','HR','IT','SALES'],'salary':[30000,50000,35000,60000,40000]}
df=pd.DataFrame(data)
res=df.groupby('department')['salary'].sum()
print(res)
```

```python
data={'department':['HR','IT','HR','IT','SALES'],'salary':[30000,50000,35000,60000,40000]}
df=pd.DataFrame(data)
res=df.groupby('department')['salary'].mean()
print(res)
```

```python
df1=pd.DataFrame({'id':[1,2,3],'name':['Alice','Bob','Charlie']})
df2=pd.DataFrame({'id':[1,2,4],'salary':[50000,60000,70000]})
res=pd.merge(df1,df2,on='id')
print(res)
```

```python
df1=pd.DataFrame({'name':['Alice','Bob','Charlie']},index=[1,2,3])
df2=pd.DataFrame({'age':[25,30,35]},index=[1,2,3])
res=df1.join(df2)
print(res)
```
